```python
import pandas as pd
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import GridSearchCV
from sklearn.preprocessing import StandardScaler
```

```python
dataSet = pd.read_csv("insurance_pre.csv")
```

```python
dataSet = pd.get_dummies(dataSet, drop_first = True, dtype = int)
dataSet
```

```python
independent = dataSet[['age', 'bmi', 'children','sex_male', 'smoker_yes']]
dependent = dataSet['charges']
```

```python
# Define parameter grid
param_grid = {'criterion':['mse','mae','friedman_mse'],
'max_features': ['auto','sqrt','log2'],
'splitter':['best','random']}
```

```python
# Set up GridSearchCV
gridSearchCV = GridSearchCV(estimator=DecisionTreeRegressor(), param_grid=param_grid, refit=True, cv=5, n_jobs=-1, verbose=3)
```

```python
gridSearchCV.fit(independent, dependent)
```

```python
#print best parameter after tuning
print(gridSearchCV.best_params_)
```

```python
age_input=float(input("Age:"))
bmi_input=float(input("BMI:"))
children_input=float(input("Children:"))
sex_male_input=int(input("Sex Male 0 or 1:"))
smoker_yes_input=int(input("Smoker Yes 0 or 1:"))
```

```python
Future_Prediction=gridSearchCV.predict([[age_input,bmi_input,children_input,sex_male_input,smoker_yes_input]])
#change the paramter,play with it.
print("Future_Prediction={}".format(Future_Prediction))
```
