```python
# Importing the pandas Library and giving an alias as pd for that, In pandas library we have methods to read csv. Hence it required to import first
# before reading a csv file

import pandas as pd
```

```python
# Reading the input csv file using the pandas library method read_csv and assigned read data to the variable dataSet

dataSet = pd.read_csv("Salary_Data.csv")
```

```python
#From the dataSet variable which has the full data set, we are extracting only YearsExperience and assigning it to the varialbe named independent

independent = dataSet[["YearsExperience"]]
```

```python
#From the dataSet variable which has the full data set, we are extracting only Salary and assigning it to the varialbe named dependent

dependent = dataSet[["Salary"]]
```

```python
# From the sklearn library model_selection class we are importing the method train_test_split
# Calling the train_test_split method by passing the extracted varialbes independent, dependent
# the train_test_split method will return 4 outputs and we are assigning it to the varalbes x_Train, x_Test, y_Train, y_Test for our further use

from sklearn.model_selection import train_test_split 
x_Train, x_Test, y_Train, y_Test = train_test_split(independent, dependent, test_size = 0.3, random_state = 0)
```

```python
# From the sklearn.linear_model library we are importing the class LinearRegression
# creating a instance / Reference of LinearRegression class and assigned to the variable regressor
# Calling the Fit method by passing the x_Train, y_Train as inputs to the Fit method
# This is the place where learning is happening. Model creation done here.

from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(x_Train, y_Train)
```

```python
# From the learned model we are reading co efficient variable and assigned to a variable named weight
# Printing the variable weight 

weight = regressor.coef_
weight
```

```python
# From the learned model we are reading bais/initial value/min value and assigned to a variable named bais
# Printing the variable bais 

bais = regressor.intercept_
bais
```

```python
# Giving the real test to our model to predict the results with test data i.e., x_Test
# y_Predict varialbe will have predicted result for the given input

y_Predict = regressor.predict(x_Test)
```

```python
# From the sklearn.metrics library we are importing the method r2_score
# calling r2_score method by passing the inputs as y_Test, y_predict to check our model learning capacity
# if its near to 1 then our model learned very well
# print the r_score to check the model learned capacity

from sklearn.metrics import r2_score
r_Score = r2_score(y_Test, y_Predict)
r_Score
```

```python
# Importing the pickle Library, pickle library is used to save our model
# created varilabe name fileName and assigned "finalized_model_Linear.sav" to it.

import pickle
fileName = "finalized_model_Linear.sav"
```

```python
# Calling the pickle library dump method to save our model by passing the model reference and fileName

pickle.dump(regressor, open(fileName, "wb"))
```

```python
# Calling the pickle library load method to load our model from the saved file and assigned it to loadedModel variable
# Calling the loadedModel.predict by passing the sample test value to check its prediction output and printed.

loadedModel = pickle.load(open(fileName, "rb"))
result = loadedModel.predict([[13]])
print(result)
```
